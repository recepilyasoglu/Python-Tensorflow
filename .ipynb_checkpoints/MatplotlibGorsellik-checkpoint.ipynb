```python
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 2
```

```python
(benimFigur, benimEksen) = plt.subplots() # figur oluşturma
benimEksen.plot(numpyDizisi1,numpyDizisi2, color = "#3995A8")#internette hex koduna baktık
benimEksen.plot(numpyDizisi2,numpyDizisi1,color = "#C96F23")
```

```python
(yeniFigur, yeniEksen) = plt.subplots()                    #çizgi kalınlığı
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 2, color="blue",linewidth = 1.0)
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 3, color="red", linewidth = 1.0)

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 4, color = "#C96F23", linestyle = "-.")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 5, color = "#C96F23", linestyle = ":")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 6, color = "#C96F23", linestyle = "--")

yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 7, color = "#000000", linestyle = "--", marker = "o", markersize = 4, markerfacecolor = "red")
yeniEksen.plot(numpyDizisi1, numpyDizisi1 + 8, color = "#000000", linestyle = "--", marker = "+", markersize = 4)
```
