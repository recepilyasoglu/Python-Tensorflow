```python
dataFrame = dataFrame.drop("transmission",axis=1) # bu s√ºtunu komple d√º≈ü√ºrd√ºk, sayƒ±sal deƒüer olmadƒ±ƒüƒ± i√ßin
```

```python
# x ve y'yi bulmak 
y = dataFrame["price"].values
x = dataFrame.drop("price",axis=1).values # "price" harici her≈üeyi alƒ±caz
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train) # dogru b√∂l√ºnd√º m√º diye kontrol ediyoruz
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train) # aynƒ± anda fit transform yapƒ±yoruz bu sefer
```

```python
x_test = scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x = x_train, y = y_train, validation_data=(x_test,y_test), batch_size=250 ,epochs=300)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi = model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi) # 3388 poundluk bir fark var aralarƒ±nda
```
