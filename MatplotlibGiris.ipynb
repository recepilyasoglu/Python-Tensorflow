```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #yazılan grafikleri kolayca gösterilemey olanak tanır(jupyter'e özel)
```

```python
yasListesi = [10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi = [20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi = np.array(yasListesi) # np olarak ayarlıyoruz numpy dizi üzeirnden çalışmakta fayda var
numpyKiloListesi = np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
         # x ekseni,    ,#y ekseni       ,red color
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.title("Kilo'nun Yaşa Göre Değişim Grafiği")
#plt.show() # baska scriptte calıstırırken bunu da ayzmamız gerekir ama  jupyter'de gerek yok, en tepede belirttigimiz ozellik sayesinde
```

```python
numpyDizisi1 = np.linspace(0,10,20) # 0'la 10 arasında 20 tane numara oluşturduk
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g") # g-- kesik çizgili grafik getirir, *- verilerin oldugu yerleri yıldızlarla gösterir
```

```python
plt.subplot(1,2,1) # 1 sıra olacak, 2 kolon olacak, şuan 1. grafiği çiziyorum
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2) # 1 sıra olacak, 2 kolon olacak, şuan 2. grafiği çiziyorum
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
```

```python
benimFigure = plt.figure()                  #with,height
figureAxes = benimFigure.add_axes([0.1,0.1,0.6,0.6])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni Veri İsmi")
figureAxes.set_ylabel("Y Ekseni Veri İsmi")
figureAxes.set_title("Grafik Başlığı")
```

```python
figure2 = plt.figure()

# ilk  iki argüman x ve y ekseninde grafiği yerleştiriyor
eksen1 = figure2.add_axes([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes([0.2,0.4,0.3,0.3])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlık")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen2.set_xlabel("X Ekseni Küçük")
eksen2.set_ylabel("Y Ekseni Küçük")
eksen2.set_title("Küçük Grafik Başlık")
```

```python
(benimFigure, benimEksen) = plt.subplots(nrows=1,ncols=2)
type(benimEksen)

for eksen in benimEksen:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    eksen.set_xlabel("X Ekseni")

plt.tight_layout()
#benimFigure
```

```python
yeniFigur = plt.figure()

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 2, label = "numpyDizisi ** 2") #üssü 3
yeniEksen.plot(numpyDizisi1,numpyDizisi1 ** 3, label = "numpyDizisi ** 3") #üssü 2 
yeniEksen.legend() # label'in sol üstte çıkmasını sağlıyor 
#yeniEksen.legend(loc = 3) # bu tarz lokasyon deerleri de verilebiliyor
```

```python
yeniFigur.savefig("benimfigur.png", dpi = 200) # grafigi kaydetme, dpi = kalitesini belli ediyor
```
