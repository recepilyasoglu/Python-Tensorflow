```python
y = dataFrame["Type"].values
x = dataFrame.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu")) # kaç tane kolon varsa unitse yani nöronlara genelde o yazılır
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) # 30' la 1 arasında herhangi bir sayı olabilir diyor by kural(yarısını koy genelde)
# sigmoid koyucaz, sigmoid 0 la 1 arasında değer veriyodu çıkış için sigmoid verilebilir

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1) # epochs 700 diyip overfiting yapmak istedik
```

```python
model.history.history
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu")) # kaç tane kolon varsa unitse yani nöronlara genelde o yazılır
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) # 30' la 1 arasında herhangi bir sayı olabilir diyor by kural(yarısını koy genelde)
# sigmoid koyucaz, sigmoid 0 la 1 arasında değer veriyodu çıkış için sigmoid verilebilir

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
# yukardaki val_loss yükseldiğinde yan çok açıldığında durdur -> EarlyStopping'in olayı  
earlyStopping = EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot() # iyileşme var, val_loss biraz daha mantıklı hale geldi, yine yeterli değilse dropout kullanılır
```

```python
model = Sequential()

model.add(Dense(units=30,activation="relu")) 
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation="sigmoid")) 

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayipDf = pd.DataFrame(model.history.history)
```

```python
kayipDf.plot() # daha öncekine göre daha iyi durumda, overfittingin üstesinden geldik
```

```python
tahminlerimiz = (model.predict(x_test) > 0.5 ).astype("int32")
```

```python
tahminlerimiz # 0->zararsız, 1->zararlı website
```

```python
# sınıflandırmamızın ne kadar düzgün sonuçlar verdiğini değerlendirek için oluşturulan matris
from sklearn.metrics import classification_report, confusion_matrix 
```

```python
print(classification_report(y_test,tahminlerimiz)) # y_test'le tahminlerimizi kıyasladık
```

```python
print(confusion_matrix(y_test,tahminlerimiz)) # y_test'le tahminlerimizi kıyasladık
# 7 tane değeri yanlış bulmuş, geri kalanları doğru bulmuş özellikle testlerimiz içerisinde 
```
